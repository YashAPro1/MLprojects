```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('heart.csv')
```

```python
df = pd.DataFrame(df)
```

```python
df.head()
```

```python
df_Sex_dumm = pd.get_dummies(df.Sex)
```

```python
df_Sex_dumm
```

```python
df_ChestPainType_dumm = pd.get_dummies(df.ChestPainType)
df_ChestPainType_dumm
```

```python
df_RestingECG_dumm = pd.get_dummies(df.RestingECG)
df_RestingECG_dumm
```

```python
df_ExerciseAngina_dumm = pd.get_dummies(df.ExerciseAngina)
df_ExerciseAngina_dumm
```

```python
df_ST_Slope_dumm = pd.get_dummies(df.ST_Slope)
df_ST_Slope_dumm
```

```python
df = pd.concat([df,df_ST_Slope_dumm,df_ExerciseAngina_dumm,df_ChestPainType_dumm,df_RestingECG_dumm,df_Sex_dumm],axis='columns')
```

```python
df
```

```python
df =df.drop(['Sex',"ChestPainType","ST_Slope","ExerciseAngina","RestingECG"],axis='columns')
```

```python
df.head()
```

```python
y=df.HeartDisease
```

```python
X = df.drop('HeartDisease',axis='columns')
```

```python
X.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(X)
```

```python
x_scaled
```

```python
from sklearn.svm import SVC
model = SVC()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x_scaled,y,test_size=0.3,random_state=30)
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
from sklearn.linear_model import LogisticRegression
modelL = LogisticRegression()
```

```python
modelL.fit(X_train,y_train)
```

```python
modelL.score(X_test,y_test)
```

```python
#looks like svm performing better
```
