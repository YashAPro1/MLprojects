```python
import pandas as pd
from sklearn.datasets import load_digits
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
```

```python
d = load_digits()
```

```python
d.feature_names
```

```python
d.target_names
```

```python
df  = pd.DataFrame(d.data,d.target)
df.head()
```

```python
df['target'] = d.target
```

```python
df
```

```python
X = df.drop('target',axis='columns')
y = df.target
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
scv_rdf_model = SVC(kernel='rbf')
```

```python
scv_rdf_model.fit(X_train,y_train)
```

```python
scv_rdf_model.score(X_test,y_test)
```

#linaer

```python
scv_linear_model = SVC(kernel='linear')
```

```python
scv_linear_model.fit(X_train,y_train)
```

```python
scv_linear_model.score(X_test,y_test)
```
