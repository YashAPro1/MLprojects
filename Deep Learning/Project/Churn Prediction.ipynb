```python
import tensorflow as tf
from tensorflow import keras
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
print("You are ready ro go!")
```

```python
df = pd.read_csv("Churn_Modelling.csv")
```

```python
df.sample(10)
```

```python
df.shape
```

```python
for columns in df:
    print(f"{columns}: {df[columns].unique()}")
```

```python
df1 = df.drop(['RowNumber','CustomerId','Surname'],axis='columns')
```

```python
df1.head()
```

```python
df1.isnull().any()
```

```python
df1['Gender'].replace({"Female":0,'Male':1},inplace=True)
```

```python
df1.sample(10)
```

```python
df2 = pd.get_dummies(df1['Geography'])
```

```python
df2 = pd.concat([df1,df2],axis="columns")
```

```python
df3 = df2.drop('Geography',axis='columns')
```

```python
df3.sample(10)
```

```python
not_scaled = ['CreditScore',"Age",'Balance','NumOfProducts','EstimatedSalary']

from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler()
df3[not_scaled] = sc.fit_transform(df3[not_scaled])
```

```python
tenure_exit_yes = df3[df3.Exited == 1].Tenure
tenure_exit_no = df3[df3.Exited == 0].Tenure
```

```python
plt.hist([tenure_exit_no,tenure_exit_yes],color=['green','red'])
```

```python
df3.sample(10)
```

```python
X = df3.drop('Exited',axis='columns')
y = df3.Exited
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=6)
```

```python
X.shape
```

```python
import tensorflow as tf
from tensorflow import keras
model = keras.Sequential([
    keras.layers.Dense(100,input_shape=(12,),activation='relu'),
    keras.layers.Dense(20,activation='relu'),
    keras.layers.Dense(15,activation='relu'),
     keras.layers.Dense(1,activation='sigmoid'),
])

model.compile(optimizer='adam',
             loss='binary_crossentropy',
             metrics=['accuracy'])

model.fit(X_train,y_train,epochs=100)
```

```python
#so we pretty much getting the same accuracy every time
```

```python
yp = model.predict(X_test)
```

```python
ypred = []
for x in yp:
    if x[0]>0.5:
        ypred.append(1)
    else:
        ypred.append(0)
```

```python
ypred[:5]
```

```python
y_test[:5]
```

```python
from sklearn.metrics import classification_report,confusion_matrix

print(classification_report(y_test,ypred))
```

```python
import seaborn as sn

cm = tf.math.confusion_matrix(labels=y_test,predictions=ypred)

plt.figure(figsize=(10,7))
sn.heatmap(cm,annot=True,fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
